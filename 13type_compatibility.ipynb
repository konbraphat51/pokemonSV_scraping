```python
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_binary

driver = webdriver.Chrome()
```

```python
#compatibilities[ATTACK][DEFEND] = COEFFICIENT
compatibilities={}
```

```python
import json

#read json
f = open("translation/type.json", "r", encoding="utf-8")
data = f.read()
f.close()

types = json.loads(data)
```

```python
#get type-id from Japanese Name
def getID(jp):
    for type_id in types:
        if jp == types[type_id]["JpName"]:
            return int(type_id)

    #couldn't find
    return -1

#scrap
for type_id in types:
    en = types[type_id]["EnName"].lower()
    jp = types[type_id]["JpName"]
    
    #get page
    driver.get("https://motemen.github.io/pokemon-type-effects/?d=%s"%en)
    html = driver.page_source
    soup = BeautifulSoup(html)

    #initialize
    compatibilities[type_id] = {}

    #set default as x1
    for num in range(18):
        compatibilities[type_id][num] = 1
    
    #get other than x1
    for defend_div in soup.find("div", class_="w-full max-w-md").find_all("section", class_="w-full")[1].find_all("div"):
        jpName = defend_div.find("span", class_="grow").text
        coef = defend_div.find("span", class_="w-8").text

        try:
            coef = int(coef)
        except:
            #coef = Â½
            coef = 0.5

        compatibilities[type_id][getID(jpName)] = coef
```

```python
#save
import json

f = open("data/compatibilities.json", "w", encoding="utf-8")
json.dump(compatibilities, f, ensure_ascii=False)
f.close()
```
